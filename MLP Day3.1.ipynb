```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
print(y)
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("Sinevalue")
plt.title("Sinewave")
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel("angle")
plt.ylabel("Cosvalue")
plt.title("Coswave")
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number Of People')
plt.show()
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.2f%%')
plt.show()
```

```python
x=np.linspace(0,10,50)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='m')
ax.scatter(x,z,color='k')
plt.show()
```

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
